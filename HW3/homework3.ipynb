```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/IRIS.csv")
```

```python
fig, axes = plt.subplots(1,2, figsize = (13,6))
sns.scatterplot(ax = axes[0], data = df, x = df["sepal_length"], y = df["petal_width"], hue = "species", palette = ["blue", "green", "red"])
sns.scatterplot(ax = axes[1], data = df, x = df["sepal_width"], y = df["petal_length"], hue = "species", palette = ["blue", "green", "red"])
```
